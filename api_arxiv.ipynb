```python
import os
import logging
import matplotlib.pyplot as plt
import pandas as pd
import json
from pandas import json_normalize
import pandas
import requests
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
url_sum= "https://api.rxivist.org/v1/authors"
#url_sum= "https://api.biorxiv.org/sum/m"
#url = "https://api.biorxiv.org/pub/2017-08-21/2021-05-20"
r = requests.get(url_sum)
x = r.json()
#df = pd.DataFrame(x['bioRxiv content statistics'])
df = pd.DataFrame(x['results'])
df.head()
display(df)
```

```python
df.info()
```

```python
var = 'new_papers'
data = pd.concat([df[var], df['month']], axis=1)
f, ax = plt.subplots(figsize=(16, 8))
plt.xticks(rotation=90)
fig = sns.barplot(x='month', y=var, data=data)
fig.axis(ymin=0, ymax=8000);
```
